```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(42)
n=20000
custid = np.arange(1,n+1)
name = [f'Cust_{i}' for i in custid]
age = np.random.randint(18,70,n)
gender = np.random.choice(['Male','Female','Others'],n)
city = np.random.choice(['Blr','Mumbai','Chennai','Hyd'],n)
sign_up = pd.date_range(start='01-01-2024',periods=n,freq='H')
income = np.random.randint(10000,1000000,n)
loyal = np.random.uniform(1,10,n)
email = [f'{i.lower()}@gmail.com' for i in name]
phone = np.random.randint(6000000000, 9999999999, n, dtype=np.int64)
```

```python
customers = pd.DataFrame({'Customer_ID':custid,
                          'Customer_name':name,
                          'Age':age,
                          'Gender':gender,
                          'city':city,
                          'signup_date':sign_up,
                          'income':income,
                          'Loyalty_points':loyal,
                          'Email_ID':email,
                          'Phone':phone})
```

```python
customers.loc[np.random.choice(n,30),'Email_ID']=np.nan
```

```python
customers.loc[np.random.choice(n,30,replace=False),'Phone']=np.nan
```

```python
pd.concat([customers,customers.sample(500)])
```
